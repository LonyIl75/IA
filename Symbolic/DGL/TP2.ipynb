# RDF graph To CSV_STR
Dans cette partie nous allons définir les fonctions qui serviront à transformer un fichier encodant un graph rdf ( format : ttl ,xml , ect ) en fichier csv 

```python
#Serialize le graph dans le format specifier : (ttl , xml , json-ld ,ect )
#src : https://rdflib.readthedocs.io/en/stable/apidocs/rdflib.html#rdflib.Graph.serialize
def graphRDF_toFormat (graph , _format  ) :
    return graph.serialize(format=_format)

# Serialize le graph en json-ld 
#NB : il existe une transformation json-ld -> csv 
def graphRDF_toJson ( graph ) :
    return graphRDF_toFormat (graph , 'json-ld' )

#Convertie le graph en json-ld -> csv    
def graphRDF_toCsv( graph ,_sep=';'  ) :
    buff = graphRDF_toJson ( graph) 
    df = pd.read_json(buff)
    return df.to_csv( sep=_sep,index=False)
```

```python
#Fonction utilitaire pour ecrire un fichier 
def concat_extension (filename , ext ):
    return  filename + "." + ext

def strFormat_toFile(buff , filename , ext =None  ):
    if not(ext is None):
        filename = concat_extension (filename , ext )
    with open( filename, "w", encoding="utf-8") as text_file:
        text_file.write(buff)
    return filename
 
def verifyExt(filename ,ext):
    if not(ext is None): 
        if not(filename.endswith(ext) ):
            filename.replace(".", "_" )
            filename = concat_extension (filename , ext )
            return filename 
    return None 
```

```python
#Ecrit un csv a partir d'un graph 
def graphRDF_toCsvFile (graph, filename ) :
    filename =verifyExt(filename , "csv")
    buff = graphRDF_toCsv(graph)
    strFormat_toFile(buff , filename )
    return filename 

#Initialise un graph a partir d'un fichier (format : ttl , xml , ect )
#src : https://rdflib.readthedocs.io/en/stable/apidocs/rdflib.html#rdflib.Graph.parse
def file_toGraph(filename , _format ):
    if (_format is None ):
        _format = os.path.splitext(filename)
    g = Graph()
    g.parse(filename , _format)    
    return g 

#Ecrit un csv a partir d'un fichier encodant un graph (format : ttl , xml , ect )
def fileRDF_toFileCSVRDF (file_graph , file_out ,_format=None):
    if _format is None :
        _format = os.path.splitext(file_graph)[1]    
    graph=file_toGraph(file_graph , _format )
    return graphRDF_toCsvFile (graph, file_out)
    
```

```python
from ast import literal_eval

#On convertie chaque element du csv en un objet json 
def dfToJson (_df ):
    df = _df.copy()
    for col_id in df.columns[1:]:
        df[col_id].fillna('{}', inplace=True) #remplace les Nan par {} pour que l'ensemble du df soit json 
        df[col_id] = df[col_id].apply(literal_eval)
    return df 
```

```python

#Classe qui va permettre de convertir id_url -> id_int
class AutoIncrementID:
    def __init__(self):
        self.counter = 0

    def get_id(self):
        self.counter += 1
        return self.counter

class StringIDMapper:
    def __init__(self):
        self.auto_increment_id = AutoIncrementID()
        self.string_to_id = {}

    def map_string_to_id(self, string):
        if string not in self.string_to_id:
            self.string_to_id[string] = self.auto_increment_id.get_id()
        return self.string_to_id[string]
    
def nodeToId(_df , mapperID = None ,id_col ='@id' ):
    if(mapperID is None ):
        mapperID = StringIDMapper()
        
    for i in range(len(_df[id_col])):
        mapperID.map_string_to_id(_df[id_col][i])
    return  mapperID 

```

```python
#Fonction utilitaire :

#Transforme un URL en label , := label par defaut 
# NB: nous avons remarqué que toutes les relations ne possédait pas forcément un label d'ou cette fonction
def url_toLabel (url ):
    p = re.compile("(.+(?:#|\/)+)")
    result =  p.search(url)
    if bool(result ):
            len_res = len(result.group(1))
            buff = url[len_res:]
            if bool(buff):
                buff = buff.replace("_"," ").split(' ', 1)
                if (len(buff)<2):  
                    buff= re.sub('([A-Z]+)', r'_\1',buff[0]).lower()
                    buff = buff.replace("_"," ").split(' ', 1)
                if(len(buff)<2):
                    buff = buff[0]
                else:
                    buff= buff[1]
            return buff

                    

```

```python
#Retourne le label de l'entité json row si ce label n'existe pas créer un label par défaut et le retourne 

def get_label(row , label = "http://www.w3.org/2000/01/rdf-schema#label",lang=["en","fr"]):
     for i in range(len(lang)):
        for text in row[label]:
            if text['@language'] == lang[i] : 
                buff= text['@value']
                buff2= buff.split(' ', 1)
                if(len(buff2)<2):
                    buff2= buff.split('_', 1)
                return buff2[1]
     buff=row["@id"]
     return url_toLabel (buff)

    
#Retourne la scope note du commentaire de l'entité json row 
#NB : nous avons remarqué que la scope note pouvait être orthographié différement ou implicite 
def get_comment(row , comment = "http://www.w3.org/2000/01/rdf-schema#comment" , lang = ["fr","en"]):
    for i in range(len(lang)):
        for text in row[comment]:
            if text['@language'] == lang[i]:
                buff= text['@value']
                p = re.compile("^ *(scope note|Scope Note|Scope note|scope Note) *: *")
                result =  p.search(buff)
                if bool(result ):
                    len_res = len(result.group(1))
                    buff = buff[len_res:]
                    return buff
                
                p = re.compile("^ *(Example|Exemple)s? *: *") #not ?exemple => scope note implicite 
                result =  p.search(buff)
                if not(bool(result )):
                    return buff
                #else:
                    #print("pas buff "+ buff)
                
 #Retourne la description dans un fichier json format (id;description)
def get_description( id  ,file_json ):
    with open(file_json, 'r' ,encoding="utf-8") as f:
      data = json.load(f)
    for item in data:
        if(item['id'] == id ):
            return item['description']
```

# Embedding :
Dans cette partie nous définirons un modèle qui nous permettra d'effectuer l'embedding necessaire aux tris des données (voir fonction : df_arrayToSingle ) et le calcul de leurs features .

```python
import torch.nn.functional as F
import torch
from transformers import BertModel, BertTokenizerFast

#Fonction utilitaires : 
def mean_pooling(model_output, attention_mask):
    token_embeddings = model_output[0] 
    input_mask_expanded = attention_mask.unsqueeze(-1).expand(token_embeddings.size()).float()
    return torch.sum(token_embeddings * input_mask_expanded, 1) / torch.clamp(input_mask_expanded.sum(1), min=1e-9)


def cosine(u, v):
    return np.dot(u, v) / (np.linalg.norm(u) * np.linalg.norm(v))
```

```python
tokenizer = BertTokenizerFast.from_pretrained("setu4993/LaBSE")
model_bert = BertModel.from_pretrained("setu4993/LaBSE")
model_bert = model_bert.eval()
```

```python
def encode_listSentences(sentences ) :
    
    encoded_input = tokenizer(sentences, padding=True, truncation=True, return_tensors='pt')
   
    with torch.no_grad():
        model_output = model_bert(**encoded_input)

    sentence_embeddings = mean_pooling(model_output, encoded_input['attention_mask'])

    sentence_embeddings = F.normalize(sentence_embeddings, p=2, dim=1)

    return sentence_embeddings


def compute_dictSimilarities ( sentence_embeddings ):
    len_list = len(sentence_embeddings)
    dict_sim = dict( ( i , []) for i in range(len_list)) #dictionnary sim by sentence
    for i in range(len_list):
        for k in range(len_list):
            if(k==i ):
                dict_sim[k].append(1.0)
                continue
            sim = cosine(sentence_embeddings[i], sentence_embeddings[k])
            dict_sim[k].append(sim)
    return dict_sim

def compute_listMeanAndVar(_dict):
    list_stat=[]
    for l in _dict.values() :
        var = np.var(l)
        mean = np.mean(l)
        list_stat.append([mean,var])
    return list_stat

        
def select_mostRepresentative(list_stat , delta_mean = 0.01  ,delta_var =0.0) :       
    id_pair = 0 
    max_pair= list_stat[id_pair]
    for i in range( id_pair +1, len (list_stat)): 
        e = list_stat[i]
        if e[0] > max_pair[0] + delta_mean :
            max_pair= e
            id_pair = i 
        elif e[0] >= max_pair[0] :
            if e[1] + delta_var < max_pair[1] :
                max_pair= list_stat[i]
                id_pair = i 
    return id_pair,max_pair 

def select_mostRepresentativeSentence(sentences ):
    if(len(sentences)>1):
        sentence_embeddings = encode_listSentences(sentences )
        dict_sim = compute_dictSimilarities ( sentence_embeddings ) 
        list_meanVar = compute_listMeanAndVar( dict_sim)
        idfinal , _ = select_mostRepresentative(list_meanVar ) 
        return sentences[idfinal],idfinal
    else :
        return sentences[0],0
```

```python
def removeDuplicate(df,id_diffs):
    pte=df.copy()
    pte=pte.reset_index(drop=True)
    mask = pte.id.duplicated(keep=False)
    grp=pte[mask].groupby("id").groups

    lst_conflict=[]
    lst_duplicate=[]

    for e in grp.values() :
        for v in range(1,len(e)) :
            idx=0
            for _id in id_diffs :
                if(pte[_id].loc[e[0]] != pte[_id].loc[e[v]] ) :
                    lst_conflict+=e
                idx+=1
            if(idx== len(id_diffs)):
                lst_duplicate+=[e[v]]

    n_pte = pte.drop(pte.index[lst_duplicate], inplace=False) 
    return n_pte,lst_conflict
```

```python

def getTrain_size(lst, percent_test = 0.1):
    return round(len(lst) * (1-percent_test))

def get_Test_Train_lst(lst , percent_test = None ):
    if (percent_test is None ):
        sz = getTrain_size(lst)
    else :
        sz = getTrain_size(lst,percent_test)
    lst_train = lst[:sz] 
    lst_test = lst[sz:]
    return lst_test ,  lst_train

def generate_RandomIds(len_lst_ids):
    ids = np.arange(len_lst_ids)
    return  np.random.permutation(ids)

    
def get_Test_Train_edges (u,v,percent_test=None ) : 
    lst_ids_edges=generate_RandomIds(len(u))
    test_ids ,train_ids= get_Test_Train_lst(lst_ids_edges,percent_test)
        
    test_u, test_v  = u[test_ids], v[test_ids]
    train_u, train_v= u[ train_ids], v[ train_ids]
    return {"test":( test_u, test_v), "train":(train_u, train_v)}#test first 
    
def g_getPos_Test_Train_edges (graph,percent_test = None ):
    pos_dict = {"test":[],"train":[]}
    for etype in hetero_graph.canonical_etypes:
        src, dst = hetero_graph.all_edges(etype=etype, order="eid")
        dict_tesTrain =  get_Test_Train_edges ( src, dst ) 
        pos_dict["test"].append({etype : dict_tesTrain["test"] })
        pos_dict["train"].append({etype :  dict_tesTrain["train"] })
    return pos_dict
    

    
```

```python
def list_equivalent (lst_1 , lst_2):
    if(len(lst_1) != len(lst_2) ):
        return False 
    for e in lst_1 :
        if not(e in lst_2):
            return False
    return True
            
def lists_equivalent( lst_lst_n):
    for lst in lst_lst_n[1:] :
        if( not( list_equivalent(lst_lst_n[0],lst))):
            return False 
    return True
```

```python
def pos_lst_edges (frames_edges,cols_src="src",cols_dst="dst"):
    pos_edges_lst=[]
    for edges_lst in frames_edges :
        pos_dict = get_Test_Train_edges ( edges_lst[cols_src],edges_lst[cols_dst],None)
        pos_edges_lst.append(pos_dict )
    return pos_edges_lst

def neg_lst_edges (frames_edges,cols_src="src",cols_dst="dst"):
    neg_edges_lst=[]
    for edges_lst in frames_edges :
        neg_dict = get_Test_Train_edges (edges_lst[cols_src],edges_lst[cols_dst],0.0)
        neg_dict["test"] = get_Test_Train_edges (edges_lst[cols_src],edges_lst[cols_dst],1.0)["test"]
        neg_edges_lst.append(neg_dict )
    return neg_edges_lst


```

```python
import copy 

def util_find (df,r,col_id1 , col_id2,col_id3,default_type=0):
    res= df[df[col_id1] == r[col_id2]]
    
    if(not(res.empty)):
        return res [col_id3].values[0]
    else :
        return 0
    
#récupère le triplet (type_1;type_rel;type_2) et ajoute à l'index triplet l'element  ([r['src']] , [r['dst']])
def htg_dict_tuple(ids_nodes_data,edges_data_type):
    dict_tuple = dict()
    for i,r in edges_data_type.iterrows():
        sujet_1 = str(util_find (ids_nodes_data,r,'id',"src",'type')) 
        sujet_2 = str(util_find (ids_nodes_data,r,'id',"dst",'type'))
        tmp_tuple=( sujet_1 ,'_'.join([sujet_1,sujet_2,str(r['rel'])]), sujet_2)
        if tmp_tuple in dict_tuple :
            dict_tuple[tmp_tuple][0].append(r['src'])
            dict_tuple[tmp_tuple][1].append(r['dst'])
        else:
            dict_tuple.update({tmp_tuple:([r['src']] , [r['dst']])})
    return dict_tuple


def resetIdx_Tuple(k ,v,rename_dicts,dict_tuple2):
    for z in [0,1]:
                _tuples=dict_tuple2[k][z]
                for i in range(len(_tuples)):
                    if not(k[z*2] in rename_dicts ):
                        rename_dicts.update({k[z*2] :StringIDMapper()})
     
                    _tuples[i]= rename_dicts[k[z*2]].map_string_to_id( str(_tuples[i]))-1

  
#remplace les elements de chaque liste d'ids dict((type_1;type_rel;type_2))[0|1] par son ordre d'apparition 
def resetIdx_dictTuple(dict_tuple, rename_dicts = None ) :
    if (rename_dicts is None ):
        rename_dicts = dict()
    dict_tuple2=copy.deepcopy(dict_tuple)
    for k ,v in dict_tuple.items():
        resetIdx_Tuple(k ,v,rename_dicts,dict_tuple2)
    return dict_tuple2 ,rename_dicts
            
```

```python
import torch.nn as nn

class HeteroDotProductPredictor(nn.Module):
    def forward(self, graph, h, etype):
        with graph.local_scope():
            graph.ndata['h'] = h
            graph.apply_edges(fn.u_dot_v('h', 'h', 'score'), etype=etype)
            return graph.edges[etype].data['score']
        
        
```

```python
class DotProductPredictor(nn.Module):
    def forward(self, graph, h):
        with graph.local_scope():
            graph.ndata['h'] = h
            graph.apply_edges(fn.u_dot_v('h', 'h', 'score'))
            return graph.edata['score']
```

```python
def construct_negative_graph(graph, k, etype):
    utype, _, vtype = etype
    src, dst = graph.edges(etype=etype)
    neg_src = src.repeat_interleave(k)
    neg_dst = torch.randint(0, graph.num_nodes(vtype), (len(src) * k,))
    return dgl.heterograph(
        {etype: (neg_src, neg_dst)},
        num_nodes_dict={ntype: graph.num_nodes(ntype) for ntype in graph.ntypes})

```

```python
class Model(nn.Module):
    def __init__(self, in_features, hidden_features, out_features, rel_names):
        super().__init__()
        self.sage = RGCN(in_features, hidden_features, out_features, rel_names)
        self.pred = HeteroDotProductPredictor()
    def forward(self, g, neg_g, x, etype):
        h = self.sage(g, x)
        return self.pred(g, h, etype), self.pred(neg_g, h, etype)
```

```python
import dgl.nn as dglnn

class RGCN(nn.Module):
    def __init__(self, in_feats, hid_feats, out_feats, rel_names):
        super().__init__()

        self.conv1 = dglnn.HeteroGraphConv({
            rel: dglnn.GraphConv(in_feats, hid_feats)
            for rel in rel_names}, aggregate='sum')
        self.conv2 = dglnn.HeteroGraphConv({
            rel: dglnn.GraphConv(hid_feats, out_feats)
            for rel in rel_names}, aggregate='sum')

    def forward(self, graph, inputs):
        # inputs are features of nodes
        h = self.conv1(graph, inputs)
        h = {k: F.relu(v) for k, v in h.items()}
        h = self.conv2(graph, h)
        return h
```

```python
import dgl.function as fn

def compute_loss(pos_score, neg_score):
    # Margin loss
    n_edges = pos_score.shape[0]
    return (1 - pos_score + neg_score.view(n_edges, -1)).clamp(min=0).mean()

def compute_auc(pos_score, neg_score):
    scores = torch.cat([pos_score, neg_score]).numpy()
    labels = torch.cat(
        [torch.ones(pos_score.shape[0]), torch.zeros(neg_score.shape[0])]).numpy()
    return roc_auc_score(labels, scores)

def compute_recall(pos_score, neg_score):
    scores = torch.cat([pos_score, neg_score]).numpy()
    labels = torch.cat(
        [torch.ones(pos_score.shape[0]), torch.zeros(neg_score.shape[0])]
    ).numpy()
    return recall_score(labels,scores)
```
