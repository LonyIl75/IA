```python
import pandas as pd 
import os

class ReadActions :
    MAP_SIZE = 612 
    intervals_ends = dict()
    intervals_dict = dict()
    intervals_dictRev = dict()
    filename_dfActions = "../data/obj/df-Actions.json"
    ex_idx = [0,2,3]
    lst_col = ['LOST_LIFE','LIFE','MOVE_TO','WEAPON','CHIP','MAPS']
    id_col = ['ID_GEN','ID_FIGHT','ENTITY_ID','ID_TURN']
    
    #maps = [ [0]*MAP_SIZE ,[0]*MAP_SIZE]
    
    def __init__(self,ids,nbGen , nbFght ):
        self.entities_r =dict()
        self.stats_turns = []
        self.turns_action = {}
        self.nbGeneration = nbGen
        self.nbFight = nbFght
        self.beg = 0
        ReadActions.init_intervals(ReadActions.filename_dfActions,ReadActions.ex_idx)
        self.nb_type = len(ReadActions.intervals_ends)
        self.df = pd.DataFrame(columns=[list(ReadActions.intervals_ends.keys())[i] for i in range(1,len(ReadActions.intervals_ends)) ]+
                               ReadActions.lst_col+ReadActions.id_col)
        self.beg_int = len(self.df.columns) - len(ReadActions.lst_col)-len(ReadActions.id_col)
        self.entities_ids = ids
        
    def setDF(self,df2):
        if(df2.columns == self.df.columns):
            self.df = pd.concat([self.df , df2])
            
        
    @staticmethod
    def emptyIntervalsEnd():
        ReadActions.intervals_ends.update({'BEGIN':0})
    @staticmethod
    def init_intervals(filename_dfA,ex_idx=[]):
        ReadActions.emptyIntervalsEnd()
        with open(filename_dfA, 'r') as f:
          data = json.load(f)
        interval_size = data["interval_size"]
        lst_data = [k for k in data.items()]
        for e in ex_idx :
            del lst_data[e]
        for k,v in lst_data  :
            for v1 in v.items():
                ReadActions.intervals_dict.update({v1[0]:v1[1]})
            ReadActions.intervals_ends.update({k:v1[1]})
        ReadActions.intervals_dictRev = reversedict(ReadActions.intervals_dict)
        
     
    @staticmethod
    def whileToken(str_token , lst_actions):
         for i in range(len(lst_actions)) :
            action = lst_actions[i]
            if(action[0] == ReadActions.intervals_dict.get(str_token)):
                i+=1
                return i
            
    def isEndFight(self,act_token):
        return act_token == self.intervals_dict.get("END_FIGHT")
    
    def isBeginFight(self,act_token):
        return act_token == self.intervals_dict.get("START_FIGHT")
    
    def isBeginTurn(self,act_token):
        return act_token == self.intervals_dict.get("NEW_TURN")
    

    def isEndTurn(self,act_token):
        return self.isEndFight(act_token) or self.isBeginTurn(act_token ) or act_token == self.intervals_dict.get("END_TURN")
                
    def addNewRow(self):
        lst =[]
        for i in range (self.beg_int):
            lst.append(list([[]]))
        for i in range (self.beg_int , len(self.df.columns)):
            lst.append([0])
            
        self.df.loc[self.getCurrentIDX()+1]=lst
    
        
    def addNewTurn(self,idx_turn):
        self.addNewRow()
        self.setColumn ('ID_TURN', idx_turn )
        
    def getCurrentIDX (self):
        return len(self.df)-1 
    
    def setColumn ( self , str_col ,val):
        if (str_col in self.df.columns):
            self.df.loc[self.getCurrentIDX()][str_col]= val 
            #print("ici  " + str(self.getCurrentIDX()), self.df.loc[self.getCurrentIDX()][str_col])
    
    def addToColumn(self,str_col,val,pos=0):
        if (str_col in self.df.columns):
            self.df.loc[self.getCurrentIDX()][str_col][pos]+= val  
            
    def addToColumnList(self,str_col,val,pos=0):
         return self.addToColumn(str_col , [val],pos) 
            
    def setIDs(self,idx_turn,entity_id):
        self.setColumn ('ID_GEN',self.nbGeneration)
        self.setColumn ('ID_FIGHT',self.nbFight)
        self.setColumn ('ENTITY_ID',self.getEntityID(entity_id))
        
    def getEntityID(self ,entity_id ):
        if len(self.entities_ids) > entity_id :
            return self.entities_ids[entity_id]
        return -1
        
    def getTurn ( self,lst_actions ,idx_turn ):
        entity_id = -1 
        k=ReadActions.whileToken("LEEK_TURN",lst_actions)
        act_token = lst_actions[k][0]
        if(self.isEndTurn(act_token)):
            return k , entity_id 
        else :
            self.addNewTurn(idx_turn)
        while k  < len(lst_actions):
            act_token = lst_actions[k][0]
            if(self.isEndTurn(act_token) ):
                return k , entity_id
            for i in range(1,len(ReadActions.intervals_ends)):
                interval = [list(ReadActions.intervals_ends.values())[i-1],list(ReadActions.intervals_ends.values())[i]]
                if(isInInterval(act_token , interval )) :
                    '''if (entity_id !=-1 and entity_id != lst_actions[k][1] ):
                        break
                    else :
                        entity_id = lst_actions[k][1]
                    '''
                    entity_id = lst_actions[k][1]
                    self.addToColumnList (list(ReadActions.intervals_ends.keys())[i],act_token)
                    self.setIDs(self.getCurrentIDX (),entity_id)
                    if(list(ReadActions.intervals_ends.keys())[i] == "Times"):
                        if(ReadActions.intervals_dictRev.get(act_token) == "MOVE_TO"):
                            lst_mv = lst_actions[k][3]
                            self.setColumn ('MOVE_TO',[lst_mv[0],len(lst_mv),lst_mv[-1]])
                            self.setColumn ('MAPS',lst_mv[:])
                            
                            '''assert(entity_id >=0 and entity_id<2)
                            for z in lst_mv :
                                self.maps[entity_id][z]+=1'''
                            
                    elif(list(ReadActions.intervals_ends.keys())[i] == "Buffs"):
                        if(ReadActions.intervals_dictRev.get(act_token) == 'HEAL'):
                            self.addToColumn('LIFE',lst_actions[k][2])
                        elif(ReadActions.intervals_dictRev.get(act_token) in ['LOST_LIFE','NOVA_DAMAGE','DAMAGE_RETURN','LIFE_DAMAGE','POISON_DAMAGE',' AFTEREFFECT']):
                            self.addToColumn('LOST_LIFE',lst_actions[k][2])
                    elif(list(ReadActions.intervals_ends.keys())[i] == "Effects"):
                        if(ReadActions.intervals_dictRev.get(act_token) == 'ADD_WEAPON_EFFECT'):
                            self.addToColumn('WEAPON',lst_actions[k][5])
                        elif(ReadActions.intervals_dictRev.get(act_token) == 'ADD_CHIP_EFFECT'):
                                self.addToColumn('CHIP',lst_actions[k][5])
                    
                    
            k+=1

                            
                    
    def getFight ( self,lst_actions ):
        i=ReadActions.whileToken("START_FIGHT",lst_actions)
        k=i
        idx_turn = 0 
        while k < len(lst_actions):
            act_token =lst_actions[k][0]
            if(self.isEndFight(act_token )):
                return k 
            nbReads , ent_id = self.getTurn(lst_actions[k:],idx_turn)
            k+=nbReads
            act_token =lst_actions[k][0]
            if(self.isBeginTurn(act_token)):
                idx_turn +=1
            k+=1
             

            
                    
```

```python
import os


def readThenWriteActions(filename_input_actions  ,entities_ids, nbGeneration , nbFight  , path_dir_input =None , path_dir_output ="..\outcome"):
    if (not ( path_dir_input is None )):
        filename_input_actions = os.path.join(path_dir_input, filename_input_actions)
    with open(filename_input_actions, 'r') as f:
      data = json.load(f)
    rdA = ReadActions(entities_ids,nbGeneration , nbFight )
    rdA.getFight(data['actions'])
    filename_csv = os.path.join(path_dir_output, '_'.join(map(str,[ nbGeneration , nbFight , round(time.time()) ]))+".csv")
    isExist = os.path.exists(path_dir_output)
    print(filename_csv)
    if not isExist:
        os.makedirs(path_dir_output)
    rdA.df.to_csv(filename_csv, sep=';', encoding='utf-8')
    return  rdA
```

```python
def recur (lst):
    res = []
    for x in lst :
        if( isinstance(x, list)):
            res += recur(x)
        else :
            res.append(x)
    return res 
def notSetCol():
    return [] 
def mergeCol (df,idx,df_res=None):
    if df_res is None :
        df_res = pd.DataFrame(columns = df.columns )
    else :
        assert(df_res.columns == df.columns)

    df_res.loc[idx]=[[]]*len(df_res.columns)
    for c in df_res.columns :
        if(not('ID' in c)):
           df_res.loc[idx][c]=recur (list(df[c]))
        else:
            df_res.loc[idx][c]=notSetCol()
    return df_res
            
```

```python
from collections import Counter

def reduceCol (df,idx , res_dict=None ):
    if res_dict is None :
        res_dict = dict.fromkeys(df.columns)
        for c in df.columns :
            res_dict[c]= list()
        
    else : 
        assert(list(res_dict.keys()) == df.columns)
        assert(isinstance(list(res_dict.values())[0],list))
    for c in df.columns :
        res_dict[c].append(dict(Counter(df[c].tolist()[0])))
    return res_dict
    
```

```python
import re 

def getGenPos():
    return 0
def getFightPos():
    return 1
def getIdsPos():
    return 2

def getDelim():
    return '_'
def getDelimIds():
    return '-'
def get_restr_ids():
    return "((?:\d+-\d+)+)"

def get_restr_idFigth(nb):
    return getDelim().join(["(\d+)"]*nb)
    
def splitIds(str_ids):
    return list(map(int , str_ids.split(getDelimIds())))


def _emptyTuple():
    return ()

def get_restr_filename_isvalid(nb=None):
    if(nb is None ):
        nb=getIdsPos()
    return "^"+getDelim().join([get_restr_idFigth(nb),get_restr_ids()])+"(?:_\d+)?$"

def b_filename_isvalid(filename):
    return filename_isvalid(filename)!=_emptyTuple()

def filename_isvalid(filename,nb = None ):
    f_str = os.path.splitext(filename)[0]
    res = re.fullmatch(get_restr_filename_isvalid(nb),f_str )
    if(bool(res)):
        return res.group(getGenPos()+1),res.group(getFightPos()+1),res.group(getIdsPos()+1),res.group(0)
    return ()
        
```

```python
import socket
import enum


def getNoClient():
    return -1
def isServ(config):
    return config['nb_connection']!=getNoClient()

class DistantConfig :
    df_port = 5000 
    MIN_PORT = 5000
    NB_CLIENT = 1
    config_model = ["host","port","nb_connection"]
        
    def __init__(self,port = None ,host =None ):
        self.config = dict.fromkeys(DistantConfig.config_model ) 
        for i ,v in enumerate([socket.gethostname() ,  DistantConfig.df_port , DistantConfig.NB_CLIENT]):
                self.config[DistantConfig.config_model[i]]=v
        if (not(port is None) and port >= DistantConfig.MIN_PORT) :
            self.config["port"] = port 
        if( not(host is None)):
             self.config["host"] = host
            
    def getValues(self):
        return self.config.values()
    
    
    def invalidValue() :
        return -1
    def getValue(self, str_id):
        if(str_id in self.config.keys()):
            return self.config[str_id]
        else :
            return DistantConfig.invalidValue()
    
class IRWFilesActions:
    SIZE_TOKEN = 100
    NB_MIN_FUNCTS = 2
    class Token(enum.Enum):
       INIT = 0
       END = 1
        
       ERROR = -1 #Attention 
    
       READY = 2
       WAIT = 3
    def __init__(self,*args):
        assert(len(args) >= IRWFilesActions.NB_MIN_FUNCTS)
        self.functs =[]
        for f in args :
            self.functs.append(f)
    def callFuncts(self , c_cmd ,*args):
        fct =self.getFuncts(c_cmd  )
        if(bool(fct)):
            return fct(*args)
        
    def getFuncts(self,g_cmd):
        if(int(g_cmd.value) < len(self.functs)):
            return self.functs[int(g_cmd.value)]
        else:
            return None 
        
    
    
    

class ReaderFilesActions (IRWFilesActions):
    
    def __init__(self,path_input_files ,port,host):
        self.input_path = path_input_files
        self.config_serv = DistantConfig(port,host)
        self.command = IRWFilesActions( self.init_client , self.finish_client)
        self.init_cst()
        #self.init_client()
        
        
    def init_cst(self,port=None):
        self.updateFiles()
        self.prev_files = []
        
    
       
    def needToRead(self):
        return  list(set(self.files) - set(self.prev_files))
    
    def getFilesInPath(self):
        return [pos_json for pos_json in os.listdir(self.input_path) if pos_json.endswith('.json')]
    
    def updateFiles(self):
         self.files = self.getFilesInPath()
    
    def setPrev(self):
        self.prev_files = self.files
    
    def readFiles (self):
        self.setPrev()
        self.updateFiles()
        json_files = self.needToRead()
        for file in json_files :
            lst = filename_isvalid( file )[:] 
            if(bool(lst)):
                n_file = os.path.join(self.input_path,file)
                with open(n_file) as f:
                    lines = f.readlines()
                    print(lines)
                #readThenWriteActions(file ,splitIds(lst[ getIdsPos()]) , lst[getGenPos()] , lst[getFightPos()] , path_dir_input =self.input_path)
            
    def init_client(self):
        
        
        host , port ,_ = list(self.config_serv.getValues())[:]

        self.client_socket = socket.socket(socket.AF_INET,socket.SOCK_STREAM) 
        self.client_socket.connect((host, port))  

        
        data =  self.client_socket.recv(1024).decode()  # receive response

        print('Received from server: ' + data)  # show in terminal

        self.client_socket.send( data.encode())
        
        while True :
            data =  self.client_socket.recv(1024).decode()  # receive response
            if(data == "READY"):
                print("READY ICI ")
                self.readFiles()
                
            if(data==""):
                print("closed connection")
                break;
            else :
                print("pas ready " +str(data))
                
        self.finish_client()
        

    def finish_client(self):
        self.client_socket.close()  # close the connection     
        
```

```python
class WriterFilesActions :
        
    
    def __init__(self,path_output_files,validateName_funct =  b_filename_isvalid , port = None  ):
        self.output_path = path_output_files
        
        self.config_serv = DistantConfig(port=port)
        self.conn = None
        self.address = None
        #self.init_serv()
        #self.command =IRWFilesActions( self.init_serv , self.finish_serv)
        self.validateName = validateName_funct
        
    def checkValidJson(content):
        return True 
    
    def getInvalidWriteCode():
        return -1
    
    def getValidWriteCode():
        return 0 
    
    def write_file(self,filename_output , content):
        if(not(bool(self.validateName(filename_output)))):
            return WriterFilesActions.getInvalidWriteCode()
        
        n_filename_out = os.path.join(self.output_path, filename_output)
        out_file = open(n_filename_out, "w")
        assert(WriterFilesActions.checkValidJson(content))
        json.dump(content, out_file, indent = 6)
        out_file.close()
        print("WRITE")
        if( not( self.conn is None  )):
            message="READY"
            self.conn.send(message.encode())
        return WriterFilesActions.getValidWriteCode()
        
    def init_serv(self):

        host , port , nb_client = list(self.config_serv.getValues())[:]
        self.server_socket = socket.socket(socket.AF_INET,socket.SOCK_STREAM)
        
        self.server_socket.bind((host, port)) 

        self.server_socket.listen(nb_client)
        
        print("PRET A LIRE "+str(nb_client))
        
        self.conn, self.address =  self.server_socket.accept() 
    
        print("Connection from: " + str(self.address))
        
        
        message="OK"
        self.conn.send(message.encode())
        
        data =  self.conn.recv(1024).decode()  # receive response
        
        print("OK message recu : " + str(data))
        
        return self.conn , self.address
        
        
        
    def get_myHostname(self):
        return self.config_serv.getValue("host")
    
    def get_myPort(self):
        return self.config_serv.getValue("port")
    
    
        
        
    
    def finish_serv(self) :
        print("Serv close ")
        self.conn.close()
        
        
        
        
    
```
